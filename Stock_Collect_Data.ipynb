```python
import requests
from bs4 import BeautifulSoup
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
headers = {"User-Agent" : user_agent}
month = "06"
df = pd.DataFrame(columns = ["date","price_open","price_close","volume"])
df1 = pd.DataFrame(columns = ["date","price_open","price_close","volume"])
while int(month) <=12:
    url = f"https://stock.wearn.com/cdata.asp?Year=112&month={month}&kind=2330"
    res = requests.get(url = url, headers = headers)
    html = BeautifulSoup(res.text,"html.parser")
    #print(soup.prettify())
    table = html.findAll("table",{"class":"mobile_img"})[0]
    trs = table.findAll("tr")[2:]
    for index,tr in enumerate (trs):
        tds = tr.findAll("td")
        date = tds[0].text
        y, m, d = date.split("/")
        date = f"{int(y)+1911}/{m}/{d}"
        price_open = tds[1].text.replace("\xa0","")
        price_close = tds[4].text.replace("\xa0","")
        volume = tds[5].text.replace("\xa0","")
        new_row = [date,price_open,price_close,volume]
        df1.loc[index] = new_row
    df = pd.concat([df, df1], ignore_index=True)
    month = str(int(month) + 1 )
    if len(month) < 2:
        month = "0"+ month
    else:
        pass
#display(df) 
df.sort_values("date", ignore_index=True)
```
